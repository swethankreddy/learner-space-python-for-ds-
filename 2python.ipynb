```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("videogamesales1.csv")
```

```python
data.head()
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df['global_sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']
```

```python
df.head(3)
```

```python
df.sort_values(by='global_sales', ascending=False, inplace=True)
```

```python
df.head()
```

```python
data['NA_Sales'] = pd.to_numeric(data['NA_Sales'],errors='coerce')
data['EU_Sales'] = pd.to_numeric(data['EU_Sales'],errors='coerce')
data['JP_Sales']  = pd.to_numeric(data['JP_Sales'],errors='coerce')
data['Other_Sales']  = pd.to_numeric(data['Other_Sales'],errors='coerce')
```

```python
plt.title('Total Number of Copies sold of each Genre globally')
plt.xlabel('Genre')
plt.ylabel('Total copies sold')
gener_sales = df.groupby('Genre')['global_sales'].sum()
gener_sales.plot()
plt.show()
```
